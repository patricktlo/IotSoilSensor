Calculation for estimating tau (RC) value of the circuit based on the capacitance perceived by the sensors.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

R = [240e3]
clk = 8e6

C = np.linspace(100e-12, 4e-9, 200)

tau = R*C

plt.plot(C*1e9,tau*1e6)
plt.xlabel("Capacitance (nF)")
plt.ylabel("tau (us)")

plt.show()
```
